```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
import cv2
```

```python
# color transformation
im = cv2.imread('candies',cv2.IMREAD_COLOR)
# option 1
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
# option 2
# b,g,r = cv2.split(im)           # get b, g, r
# im = cv2.merge([r,g,b])     # switch it to r, g, b
# option 3
#im = mpimg.imread('banan.jpg')
imhsv = cv2.cvtColor(im,cv2.COLOR_BGR2HSV)
plt.subplot(121), plt.imshow(im)
plt.subplot(122), plt.imshow(imhsv)
```

```python
# histogram
plt.figure(1)

plt.hist(im[0].ravel(),256)
plt.hist(im[1].ravel(),256)
plt.hist(im[2].ravel(),256)
plt.show()
```

```python
# convolution
kernel = np.ones((3,3),np.float32)/9
kernel = np.ones((5,5),np.float32)/25
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(im,-1,kernel)

plt.figure(1,figsize=(10,20))
plt.subplot(121),plt.imshow(im),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#blur = cv2.blur(im,(5,5))
blur = cv2.GaussianBlur(im,(5,5),0)

plt.figure(1,figsize=(10,20))
plt.subplot(121),plt.imshow(im),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
